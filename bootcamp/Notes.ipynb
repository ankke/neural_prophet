* Trend: In which there is no fixed interval and any divergence within the given dataset is a continuous timeline. The trend would be Negative or Positive or Null Trend
* Seasonality: In which regular or fixed interval shifts within the dataset in a continuous timeline. Would be bell curve or saw tooth
* Cyclical: In which there is no fixed interval, uncertainty in movement and its pattern
* Irregularity: Unexpected situations/events/scenarios and spikes in a short time span.

- **traditional** vs. **deep-learning** models
- **one-step** vs. **multi-step** prediction
- **single** vs. **multiple time** series
- **local** vs. **global** model

- **trend** - upward/down trend, predictable
- **seasonality** - fixed time interval, repetitive pattern, predictable, number of Fourier transforms determines fit
- **events** - spikes in a short time span, high fluctuation
- **lagged regressors** - Add lagging features if the model is slow to react to sudden data distribution changes. 
- **auto regression** - An auto-regressive model is a simple model that predicts future performance based on past performance. It is mainly used for forecasting when there is some correlation between values in a given time series and the values that precede and succeed (back and forth).

**Stationary TS** The mean value of them should be completely constant in the data during the analysis.
The variance should be constant with respect to the time-frame.
Covariance measures the relationship between two variables.

```python
import pandas as pd
from neuralprophet import NeuralProphet, set_log_level
```

```python
data_location = "../../neuralprophet-data/datasets/"

sf_load_df = pd.read_csv(data_location + "births_us.csv")
set_log_level("ERROR")
```

# Generic forecast: Time-based features only

```python
m = NeuralProphet(
    weekly_seasonality=3,
    daily_seasonality=3,
    trend_reg=1,
)
df_train, df_test = m.split_df(sf_load_df, freq="auto", valid_p=0.1)

metrics = m.fit(df_train, freq="auto", validation_df=df_test, progress="plot")
```

```python
metrics.tail(1)
```

```python
forecast = m.predict(df_train)
fig = m.plot(forecast, plotting_backend="plotly")
fig
```

```python
fig.show()
```

```python
forecast = m.predict(df_test)
m = m.highlight_nth_step_ahead_of_each_forecast(1)
fig_test = m.plot(forecast)
fig_test
```

```python
fig_test.show()
```

```python
fig_param = m.plot_parameters()
fig_param
```

# 1 step ahead forecast with AR-Net: Using a Neural Network


```python
m = NeuralProphet(
    n_lags=1,
    num_hidden_layers=4,
    d_hidden=32,
    learning_rate=0.01,
)
df_train, df_test = m.split_df(sf_load_df, freq="auto", valid_p=0.1)

metrics = m.fit(df_train, freq="auto", validation_df=df_test, progress="plot")
```

```python
metrics.tail(1)
```

```python
forecast = m.predict(df_train)
fig = m.plot(forecast)
```

```python
fig.show()
```

```python
forecast = m.predict(df_test)
m = m.highlight_nth_step_ahead_of_each_forecast(1)
fig_test = m.plot(forecast)
```

```python
fig_test.show()
```

```python
fig_param = m.plot_parameters()
fig_param.show()
```
